```python
import pandas as pd
```

```python
c=pd.read_csv("customer_subscription_churn_usage_patterns.csv")
c
```

```python
c.columns[c.isna().any()]
```

```python
c.describe()
```

```python
c.info()
```

```python
c=c.drop(["signup_date"],axis="columns")
c
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
c["plan_type"]=le.fit_transform(c["plan_type"])
c["plan_type"]
```

```python
c
```

```python
le=LabelEncoder()
c["churn"]=le.fit_transform(c["churn"])
c["churn"]
```

```python
c
```

```python
c.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
x=c[["plan_type","monthly_fee","avg_weekly_usage_hours","tenure_months"]]
y=c["churn"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
x_train
```

```python
s=StandardScaler()
x_train=s.fit_transform(x_train)
x_train
```

```python
x_test=s.transform(x_test)
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric="euclidean")
knn.fit(x_train,y_train)
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
