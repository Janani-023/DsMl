```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a=np.array([1,2,3,4,5])
b=np.array([50,60,70,80,90])
plt.bar(a,b)
plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([50,60,70,80,40])
plt.plot(a,b,marker='o',ms=20)
plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```

```python
region=["west","north","east","south"]
sales=[23,27,22,29]
plt.pie(sales,labels=region,autopct='%1.1f%%',colors=["#ff0000","pink","blue","green"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

region = ["west", "north", "east", "south"]
sales = [23, 27, 22, 29]

cmap = plt.cm.viridis
colors = cmap(np.linspace(0, 1, len(sales)))

plt.pie(
    sales,
    labels=region,
    autopct='%1.1f%%',
    colors=colors
)

plt.legend()
plt.show()
```

```python
a=np.array([1,2,3,4,5])
b=np.array([50,60,70,80,90])
plt.figure(figsize=(10,8))
plt.subplot(1,2,1)

plt.bar(a,b)
plt.subplot(1,2,2)
plt.plot(a,b)

plt.title("days vs sales")
plt.xlabel("days")
plt.ylabel("sales")
plt.show()
```
